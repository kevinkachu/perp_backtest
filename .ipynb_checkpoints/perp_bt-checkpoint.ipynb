```python
import datetime as dt
import numpy as np
import pandas as pd
import pytz

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_hdf('./data_backtest_0.hdf').sort_values(by=['close_dt_pst','tsec'])
df['price_chg'] = df['close_next'] - df['close']
df['price_diff'] = df['price'] - df['close']
df['pct_diff'] = df['price_diff']/df['close']
```

```python
df
```

```python
df['pct_diff'] = df['price_diff']/df['close']


pct_buy = -0.005231
pct_flat = -0.002140
pct_sell = 0.000091
df_trades = pd.DataFrame(columns=['pos','open_time','close_time','open_price','close_price',
                                  'price_max','price_min','type'])
k = 0
df_trades.loc[k,'pos'] = 0
df_trades.loc[k,'price_min'] = df.loc[0,'price']
df_trades.loc[k,'price_min'] = df.loc[0,'price']

for i in range(len(df)):
    trade_type = None
    if int(i/10000)*10000 == i:
        print(i,dt.datetime.now())
    row = df.iloc[i]
    if row['pct_diff'] < pct_buy:
        if df_trades.loc[k,'pos'] == 0:
            trade_type = 'open_long'
        elif df_trades.loc[k,'pos'] == -1:
            trade_type = 'flip_long'
    elif row['pct_diff'] > pct_sell:
        if df_trades.loc[k,'pos'] == 0:
            trade_type = 'open_short'
        elif df_trades.loc[k,'pos'] == 1:
            trade_type = 'flip_short'
    elif df_trades.loc[k,'pos'] == 1 and row['pct_diff'] > pct_flat:
        trade_type = 'close'
    elif df_trades.loc[k,'pos'] == -1 and row['pct_diff'] < pct_flat:
        trade_type = 'close'
    df_trades.loc[k,'price_max'] = max(row['price'],df_trades.loc[k,'price_max'])
    df_trades.loc[k,'price_min'] = min(row['price'],df_trades.loc[k,'price_min'])

    if trade_type is not None:
        if trade_type in ['open_long','open_short']:
            k = k + 1
            df_trades.loc[k,'pos'] = 1 if trade_type == 'open_long' else -1
            df_trades.loc[k,'open_price'] = row['price']
            df_trades.loc[k,'open_pct_diff'] = row['pct_diff']
            df_trades.loc[k,'price_max'] = row['price']
            df_trades.loc[k,'price_min'] = row['price']
            df_trades.loc[k,'open_time'] = row['close_dt_pst'] + dt.timedelta(seconds=row['tsec'])
            df_trades.loc[k,'type'] = trade_type
        else:
            #first close and then open a new pos if we are flipping
            df_trades.loc[k,'pos'] = 0
            df_trades.loc[k,'close_price'] = row['price']
            df_trades.loc[k,'close_pct_diff'] = row['pct_diff']
            df_trades.loc[k,'close_time'] = row['close_dt_pst'] + dt.timedelta(seconds=row['tsec'])
            if trade_type in ['flip_long','flip_short']:
                k = k + 1
                df_trades.loc[k,'pos'] = 1 if trade_type == 'flip_long' else -1
                df_trades.loc[k,'open_price'] = row['price']
                df_trades.loc[k,'open_pct_diff'] = row['pct_diff']
                df_trades.loc[k,'price_max'] = row['price']
                df_trades.loc[k,'price_min'] = row['price']
                df_trades.loc[k,'open_time'] = row['close_dt_pst'] + dt.timedelta(seconds=row['tsec'])
                df_trades.loc[k,'type'] = 'open_long' if trade_type == 'flip_long' else 'open_short'
#         print(i,k,trade_type)
#     if i > 10000:
#         break
            
            
```

```python
df_trades['pnl'] = 0
df_trades['edge'] = np.abs(df_trades['open_price'] - df_trades['open_price']/(1+df_trades['open_pct_diff']))
df_trades.loc[df_trades['type']=='open_short','pnl'] = df_trades['open_price'] - df_trades['close_price']
df_trades.loc[df_trades['type']=='open_long','pnl'] = -(df_trades['open_price'] - df_trades['close_price'])
df_trades.mean()
```
